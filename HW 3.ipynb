# Question 6

```r
propertieswide <- patient_properties %>%
  pivot_wider(names_from = property, values_from = value)

head(propertieswide, 10)
```

# Question 7

```r
leftjoin <- patient_names %>%
left_join(propertieswide, by = "ID")

head(leftjoin, 10)
```

# Question 8

```r
leftjoin <- leftjoin %>%
  mutate(
    FIRST = gsub("[0-9]", "", FIRST),
    LAST  = gsub("[0-9]", "", LAST),)
head(leftjoin, 10)
```

# Question 9

```r
categorical <- leftjoin %>%
  select(-ID, -FIRST, -LAST,) %>%
  select(where(is.character))

for (col in names(categorical)) {
  cat("\n---", col, "---\n")
  counts <- categorical %>%
    count(.data[[col]], sort = TRUE)
  mdpre(counts)
}

```

```r
library(dplyr)
library(tidyr)

leftjoin <- leftjoin %>%
  mutate(
    BIRTHDATE = if_else(year(BIRTHDATE) >= 2000 & year(BIRTHDATE) <= 2099,
    BIRTHDATE %m-% years(100), 
      BIRTHDATE)
    )
```

```r
head(leftjoin, 10)
```

```r
leftjoin <- leftjoin %>%
  mutate(
      MARITAL = if_else(MARITAL %in% c("Fine", "male"), NA_character_, MARITAL),
        GENDER = if_else(GENDER == "Male", "M", GENDER),
      GENDER = if_else(GENDER == "Female", "F", GENDER),
        GENDER = if_else(GENDER == "female", "F", GENDER),    
    ETHNICITY = if_else(ETHNICITY == "hispani", "hispanic", ETHNICITY),
      ETHNICITY = if_else(ETHNICITY == "nonhispani", "nonhispanic", ETHNICITY),
    RACE = if_else(
      RACE == "asiann", "asian", RACE))
```

```r
library(tidyverse)

categorical_cols <- c("GENDER", "ETHNICITY", "RACE", "MARITAL")

for (col in categorical_cols) {
  cat("\n---", col, "---\n")
  counts <- leftjoin %>%
    count(.data[[col]], sort = TRUE)
  print(counts)
}
```

# Question 11

```r
library(dplyr)
library(tidyr)

leftjoin <- leftjoin %>%
  mutate(
    BIRTHDATE = if_else(year(BIRTHDATE) >= 2000 & year(BIRTHDATE) <= 2099,
    BIRTHDATE %m-% years(100), 
      BIRTHDATE),
      
    age = interval(BIRTHDATE, Sys.Date()) / years(1)
  ) %>%
  
  filter(!is.na(age) & !is.na(GENDER))
```

```r
cat("Birthdate range:\n")
print(range(leftjoin$BIRTHDATE, na.rm = TRUE))

cat("\nAge summary:\n")
print(summary(leftjoin$age))


ggplot(leftjoin, aes(x = age, fill = GENDER)) +
  geom_histogram(bins = 30, position = "dodge", alpha = 0.5) +
  labs(
    title = "Histogram of Ages by Gender",
    x = "Age (years)",
    y = "Count"
  ) +
  theme_minimal()
```

```r
leftjoin <- leftjoin |>
  mutate(
BIRTHDATE = as.Date(BIRTHDATE, format = "%m/%d/%y"),
BIRTHDATE = if_else(BIRTHDATE > as.Date("2000-01-01"),
BIRTHDATE - years(100),
BIRTHDATE),

age = interval(BIRTHDATE, Sys.Date()) / years(1)) %>%
filter(!is.na(age) & !is.na(GENDER))
cat("Birthdate range:\n")
print(range(leftjoin$BIRTHDATE, na.rm = TRUE))

cat("\nAge summary:\n")
print(summary(leftjoin$age))

ggplot(leftjoin, aes(x = BIRTHDATE, y = MARITAL, color = GENDER)) +
geom_jitter(alpha = 0.6, width = 0, height = 0.2) +
labs(title = "Birthdate vs. Marital Status",
    x = "Birthdate",
    y = "Marital Status",
    color = "Gender") +
  theme_minimal()
```
