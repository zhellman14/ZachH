```r
cat("a) Use read_csv the cars.csv and store it in a data frame.")
```

```r
dataframe <- read_csv("cars.csv")
```

```r
cat("b) Then, group the data frame by Make, get averages across the numeric variables, and then sort by
Volume in descending order. Hint: Use summarise(across(c(), mean)) to get the averages.")
```

```r
dataframe %>% group_by(Make) %>%
  summarise(across(where(is.numeric), mean)) %>%
  arrange(desc(Volume))
```
